```python
import string
import random

print("🔐 Welcome to the Password Generator!")

# Step 1: Ask for password length
length = int(input("Enter desired password length: "))

# Step 2: Ask which character types to include
include_uppercase = input("Include uppercase letters? (y/n): ").lower() == 'y'
include_digits = input("Include digits? (y/n): ").lower() == 'y'
include_symbols = input("Include symbols? (y/n): ").lower() == 'y'

# Step 3: Define character pools
lowercase = string.ascii_lowercase
uppercase = string.ascii_uppercase if include_uppercase else ''
digits = string.digits if include_digits else ''
symbols = string.punctuation if include_symbols else ''

# Step 4: Combine character pools
all_characters = lowercase + uppercase + digits + symbols

# Step 5: Check if at least one type was selected
if not all_characters:
    print("❌ Error: No character types selected.")
    exit()

# Step 6: Generate password
password = ''.join(random.choice(all_characters) for _ in range(length))

# Step 7: Output the result
print("\n🔑 Your Generated Password is:", password)
```
